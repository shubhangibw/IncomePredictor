```python
# STEP 0: remove features that I think are unnecessary or repetitive
list = ['fnlwgt', 'education', 'relationship']
df = df.drop(columns=list, axis=0)
df.head(10)
```

```python
print(np.sum(df.isna()))
df.dtypes
```

```python
# STEP 1: replacing the null values of 'age' and 'hours-per-week' columns with the mean
# of their respective columns
mean_ages=df['age'].mean()
mean_hours=df['hours-per-week'].mean()
df['age'].fillna(value=mean_ages, inplace=True) 
df['hours-per-week'].fillna(value=mean_hours, inplace=True)
```

```python
np.sum(df.isna())
```

```python
# STEP 2: remove the null values in the entire DataFrame
df.dropna(axis=0, inplace=True)
```

```python
np.sum(df.isna())
```

```python
# STEP 3: change the values of the label column into a binary classification of either
# '0' if it's <=50K or '1' if it's >50K
print(df['income_binary'])
df['income_binary'] = np.where(df['income_binary'] == '>50K', 1, 0)
df['income_binary']
```

```python
# STEP 4: one-hot encode the columns that include objects
list = df.select_dtypes(include=['object']).columns.tolist()
df[list].nunique()
```

```python
# included the features that had less than 15 unique values
to_encode = ['race','sex_selfID','workclass','occupation','marital-status']
```

```python
for colname in to_encode:
    df_encoded = pd.get_dummies(df[colname], prefix=colname+'_')
    df = df.join(df_encoded)
```

```python
df.drop(columns=to_encode,axis=1,inplace=True)
```

```python
df
```

```python
features=(df.select_dtypes(include=['number']))
y=df['income_binary']
X=features.drop(columns='income_binary', axis=1)

print("Number of examples: " +str(X.shape[0]))
print("\nNumber of features: "+str(X.shape[1]))
print("\nthe features: \n"+str(X.columns))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=123)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#decision tree
def train_test_DT(X_train,X_test,y_train,y_test,depth,leaf=1,crit='entropy'):
    model=DecisionTreeClassifier(criterion=crit,max_depth=depth,min_samples_leaf=leaf)
    model.fit(X_train,y_train)
    class_label_predictions=model.predict(X_test)
    acc_score=accuracy_score(y_test,class_label_predictions)
    return acc_score
```

```python
#decision tree
max_depth_range=[2,5,8,10,12,15,20,25]
acc=[]
for md in max_depth_range:
    score=train_test_DT(X_train,X_test,y_train,y_test,md)
    print('max depth='+str(md)+', acc score: '+str(score))
    acc.append(float(score))
```

```python
#decision tree
fig = plt.figure()
ax = fig.add_subplot(111)
p=sns.lineplot(x=max_depth_range,y=acc,marker='o',label='full training set')
plt.title('test set accuracy of DT predictions')
ax.set_xlabel('max depth')
ax.set_ylabel('accuracy')
plt.show()
```
